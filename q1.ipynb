Construct an image with RGB seperations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rgb_image = np.zeros((100, 150, 3), dtype=np.uint8)
# rgb_image
```

```python
rgb_image[:,0:50,0] = 255
rgb_image[:,51:100,1] = 255
rgb_image[:,101:150,2] = 255
```

```python
plt.imshow(rgb_image)
```

```python
rgb_image = rgb_image[:,:,::-1]
plt.imshow(rgb_image)
```
