OpenCV reads BGR

```python
cimg = cv2.imread(r'/home/arvind/fundamentals-of-computer-vision/my_practice/computer_vision/01_Introduction to images/tulips.jpg')
cimg = cimg[:,:,::-1]
```

```python
R = cimg[:,:,0]
G = cimg[:,:,1]
B = cimg[:,:,2]
```

```python
plt.figure(1)

plt.subplot(231)
plt.imshow(cimg)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(cimg)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(cimg)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(R, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(235)
plt.imshow(G, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(236)
plt.imshow(B, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
R[100:400, 100:400] = 255
G[100:400, 100:400] = 0
B[100:400, 100:400] = 0
cimg[:,:,0] = R
cimg[:,:,1] = G
cimg[:,:,2] = B
```
